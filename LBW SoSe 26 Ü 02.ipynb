Ziel dieser Übung ist es, die Einsteinsche Summenkonvention in Python zur Berechnung von Spannung und Dehnungen zu verwenden.

Wir nutzen in dieser Übung die Funktion `einsum` des Pakets `numpy` um Tensorberechnungen durchzuführen.

````python
import numpy as np
````

Außerdem wollen wir uns viele Tensoren *hübsch* ausgeben. Hierzu nutzen wir folgende Pakete und und eine Funktion, um Tensoren als Latex code auszugeben:

````python
from IPython.display import display, Math

def array_to_latex(A, fmt="{:.3g}", vector_as_column=True, tensor3d_axis=0):
    A = np.asarray(A)

    # Scalar (0D)
    if A.ndim == 0:
        return fmt.format(A.item())

    # Vector (1D)
    elif A.ndim == 1:
        if vector_as_column:
            body = r" \\ ".join(fmt.format(x) for x in A)
            return r"\begin{bmatrix}" + body + r"\end{bmatrix}"
        else:
            body = " & ".join(fmt.format(x) for x in A)
            return r"\begin{bmatrix}" + body + r"\end{bmatrix}"

    # Matrix (2D)
    elif A.ndim == 2:
        rows = [
            " & ".join(fmt.format(x) for x in row)
            for row in A
        ]
        body = r" \\ ".join(rows)
        return r"\begin{bmatrix}" + body + r"\end{bmatrix}"
    
    else:
        raise ValueError("Only scalars, vectors (1D), and matrices (2D) are supported.")

def printMatrix(mat, name = None):
    if name is None:
        name = ''
    else:
        name += "="
    display(Math(name + array_to_latex(mat)))
````

Grundsätzlich funktioniert die `einsum` funktion so, dass wir eine Zeichenfolge geben, die eine Gleichung in Indexnotation beschreibt.
Als weitere Parameter geben wir der Funktion alle Tensoren/Vektoren, die auf der linken Seite der Gleichung stehen:

`np.einsum('inputIndex1,inputIndex2,...->outputIndex1,outputIndex2', A, B, C, ...)`

Der Teil nach dem `->` ist dabei optional. Am besten schauen wir uns das Ganze anhand von Beispielen in Übung 1 an.
